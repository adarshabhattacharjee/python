## Digital Image Representation

In Opencv, images are not treated as pictures but as numerical data.
Each image is stored as a matrix (array) of pixel values.

- Rows represent image height
- Columns represent image width
- Each element stores pixel intensity or color information

OpenCV uses NumPy arrays to store and process images efficiently.

## Image Filtering

Filtering processes an image by modifying pixel values based on neighboring pixels.
Blurring and smoothing are commonly used to reduce noise and improve results.


```python
img=cv2.imread("D:\\opencv-workshop\\Resources-Photos\\cats.jpg")
```

```python
# Averaging
# In the kernel intensity of middle pixel is average of surrounding pixels
Average=cv2.blur(img,(5,5))
cv2.imshow('Average Blur',Average)
cv2.imshow('Cats',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Gaussian Blur
# Each surrounding pixel given a weight, center pixel is average of the weight
# Less Blurring as compared to the averaging method
```

```python
g=cv2.GaussianBlur(img,(3,3),0)
# 0 is std dev in X axis
cv2.imshow('Gauss',g)
cv2.imshow('Blur',Average)
cv2.imshow('Cats',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Median Blur(use median)(not meant for high kernel size)
m=cv2.medianBlur(img,3)
cv2.imshow('Median',m)
cv2.imshow('Gauss',g)
cv2.imshow('Blur',Average)
cv2.imshow('Cats',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Bilateral Blur
```

```python
bilateral = cv2.bilateralFilter(
    img,
    d=9,
    sigmaColor=75,
    sigmaSpace=75
)
```

```python
cv2.imshow("Original", img)
cv2.imshow("Average Blur", Average)
cv2.imshow("Gaussian Blur", g)
cv2.imshow("Median Blur", m)
cv2.imshow("Bilateral Blur", bilateral)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

#simple applicartion of contours - shape detection using contours

import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('D:\opencv-workshop\\Resources-Photos\\shapes.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, threshold = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Process each contour
for i, contour in enumerate(contours):
    if i == 0:
        continue

    # Approximate contour shape
    approx = cv2.approxPolyDP(contour, 0.01 * cv2.arcLength(contour, True), True)

    # Draw contour
    cv2.drawContours(img, [contour], 0, (0, 255, 0), 2)

    # Find center
    M = cv2.moments(contour)
    if M['m00'] != 0:
        x = int(M['m10'] / M['m00'])
        y = int(M['m01'] / M['m00'])

    # Detect shape
    sides = len(approx)
    if sides == 3:
        label = 'Triangle'
    elif sides == 4:
        label = 'Quadrilateral'
    elif sides == 5:
        label = 'Pentagon'
    elif sides == 6:
        label = 'Hexagon'
    else:
        label = 'Circle'

    # Label the shape
    cv2.putText(img, label, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,0), 2)
cv2.imshow('shapes', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Morphological Operations

```python
img = cv2.imread("D:\\opencv-workshop\\Resources-Photos\\lady.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
kernel = np.ones((5, 5), np.uint8)
```

```python
eroded = cv2.erode(binary, kernel, iterations=1)
```

```python
dilated = cv2.dilate(binary, kernel, iterations=1)
```

```python
import cv2

face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

eye_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_eye.xml"
)



cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Webcam not accessible")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray, scaleFactor=1.3, minNeighbors=5
    )

    for (x, y, w, h) in faces:
        
        cv2.rectangle(
            frame, (x, y), (x + w, y + h), (0, 255, 0), 2
        )

        
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        
        eyes = eye_cascade.detectMultiScale(
            roi_gray, scaleFactor=1.1, minNeighbors=10
        )

        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(
                roi_color,
                (ex, ey),
                (ex + ew, ey + eh),
                (255, 0, 0),
                2
            )

    
    cv2.putText(
        frame, f"Faces: {len(faces)}",
        (20, 40),
        cv2.FONT_HERSHEY_SIMPLEX,
        1, (0, 255, 0), 2
    )

    cv2.imshow("Face & Eye Detection", frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord("q") or key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
